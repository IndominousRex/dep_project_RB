```python
# %%time

import concurrent.futures
import time
import re
import pickle
import random

start = time.perf_counter()

rules = open("data/samples.txt", "r").readlines()
rules = [i[0:-1] for i in rules]

data = pickle.load(open("data/data.pkl", "rb"))


def preprocess(text):
    text = re.sub(r"@[A-Za-z0-9]+", "", text)
    text = re.sub(r"[0-9]+", "", text)
    text = re.sub(r"#[A-Za-z0-9]*", "", text)
    text = re.sub(r"https?:\/\/\S+", "", text)
    text = re.sub(r"[\n]", " ", text)
    text = re.sub(r"['\",*%$#()]", " ", text)
    text = text.encode("ascii", "ignore").decode("ascii")
    return text.strip()

data=random.sample(data,len(data))

data = [[preprocess(i[0]), i[1]] for i in data[:350]]


def pattern_search(compiled_pattern, text):
    return bool(compiled_pattern.search(text))


match_per_rule = []
k = 0

# for rule in rules[:]:
#     pattern = re.compile(rule, re.IGNORECASE)

#     for text, _ in data:
#         if pattern_search(pattern,text):
#             k+=1

for text, _ in data:
    for rule in rules:
        pattern = re.compile(rule, re.IGNORECASE)
        if pattern_search(pattern, text):
            k += 1
            break
        
    match_per_rule.append(k)
```

```python
import winreg

def get_installed_apps():
    app_list = []
    try:
        # Open the Windows Registry key where installed applications are stored
        key = winreg.OpenKey(winreg.HKEY_LOCAL_MACHINE, r"SOFTWARE\Microsoft\Windows\CurrentVersion\Uninstall")

        # Iterate through subkeys (each subkey represents an installed application)
        for i in range(winreg.QueryInfoKey(key)[0]):
            subkey_name = winreg.EnumKey(key, i)
            subkey = winreg.OpenKey(key, subkey_name)

            # Try to retrieve the DisplayName and DisplayVersion values
            try:
                display_name = winreg.QueryValueEx(subkey, "DisplayName")[0]
                display_version = winreg.QueryValueEx(subkey, "DisplayVersion")[0]
                app_list.append(f"{display_name} - {display_version}")
            except FileNotFoundError:
                pass

            winreg.CloseKey(subkey)

    except Exception as e:
        print(f"Error: {e}")

    return app_list

if __name__ == "__main__":
    installed_apps = get_installed_apps()

    if installed_apps:
        print("List of installed applications:")
        for app in installed_apps:
            print(app)
    else:
        print("No installed applications found.")

```
