```python
def preprocess(text):
    text = re.sub(r"@[A-Za-z0-9]+", "", text)
    text = re.sub(r"[0-9]+", "", text)
    text = re.sub(r"#[A-Za-z0-9]*", "", text)
    text = re.sub(r"https?:\/\/\S+", "", text)
    text = re.sub(r"[\n]", " ", text)
    text = text.encode("ascii", "ignore").decode("ascii")
    return text.strip()
```

```python
data = [[preprocess(i[0]),i[1]] for i in data]
```

```python
rules=open("rules.txt","r").readlines()
```

```python
len(rules),len(data)
```

```python
c = np.zeros((len(rules), len(data)))
i, j, k = 0, 0, 0
for rule in rules:
    j = 0
    for text, _ in data:
        if bool(re.match(rule, text,re.IGNORECASE)):
            print(f"{rule} matches {text}")
            c[i][j] += 1
            k += 1
        j += 1
    i += 1
print(f"Rule firings = {k}")
```
